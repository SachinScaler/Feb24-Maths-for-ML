```python
def caculate_derivate_at_x(func, x, delta=0.00001):
    # f(x+delta)-f(x)/delta
    return (func(x + delta) - func(x)) / delta

func = lambda x: x**2
x = 2
caculate_derivate_at_x(func, x)
```

```python
import math

func = lambda x: x**2+x**3+ math.sin(x)*math.log(x)
x = 2
caculate_derivate_at_x(func, x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(func):
  x_values = np.linspace(-10, 10, 1000)
  y_values = [func(x) for x in x_values]

  plt.plot(x_values, y_values)


func = lambda x: x**2
plot_function(func)
```

```python
def plot_derivate(func, delta = 0.0001):
    x_values = np.linspace(-10, 10, 1000)
    # delta has to be close to 0, but cannot be 0, that would give division by 0
    y_values = (func(x_values + delta) - func(x_values)) / delta

    plt.plot(x_values, y_values)

func = lambda x: x**2
plot_derivate(func)
```
